```python
%matplotlib inline
import nibabel as nib
from nibabel.processing import resample_to_output
import numpy as np
import pandas as pd
from glob import glob
import re
import matplotlib.pyplot as plt
import os
```

```python
def make_mask(les_files, overlap, resolution):
    overlap_data = None
    les_collector = []
    for les_file in les_files:
        les_img = nib.load(les_file)
        les_img_resamp = resample_to_output(les_img, voxel_sizes=(resolution))
        les_data = les_img_resamp.get_data()
        les_collector.append(les_data)  
        if overlap_data is None:
            overlap_data = les_data.copy()
        else:
            overlap_data += les_data  
    les_mask_data = overlap_data > overlap
    les_mask_data.astype(int, copy=False)
    
    les_stack_data = np.stack(les_collector, axis=-1)
    les_stack_img = nib.Nifti1Image(les_stack_data, les_img_resamp.affine, les_img_resamp.header)
    les_mask_img = nib.Nifti1Image(les_mask_data, les_img_resamp.affine, les_img_resamp.header)
    
    return les_mask_img, les_stack_img
    
```

```python
lesion_masks = glob('../data/raw/subs-405/*.nii.gz')
mask_img, les_stack_img = make_mask(lesion_masks, 5, (3,3,3))
mask_img.to_filename('../data/playground/overlap_mask_3mm.nii.gz')
les_stack_img.to_filename('../data/playground/subs_3mm.nii.gz')
```

```python
# LassoCV with pca
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import decomposition as dcm
import random
random.seed(123)
mse_data = []
rsq_data = []
coef_data = []
models = []
transforms = []
for _ in range(10):
    X_train, X_test, y_train, y_test = train_test_split(group_df.sort_index(), behav_df.sort_index()['score'], test_size=0.10)
    
    # PCA feature reduction and LASSO cross validated
    pca = dcm.PCA(n_components=0.70, svd_solver='full')
    pca_x_train = pca.fit_transform(X_train)
    reg = linear_model.LassoCV(n_jobs=20, cv=50, n_alphas=1000, verbose=True, max_iter=10000, selection='random')
    reg.fit(pca_x_train, y_train)
    
    # transform the data
    pca_x_test = pca.transform(X_test)
    y_pred = reg.predict(pca_x_test)
    
    mse = mean_squared_error(y_test, y_pred)
    rsq = r2_score(y_test, y_pred)
    mse_data.append(mse)
    rsq_data.append(rsq)
    coef_data.append(reg.coef_)
    models.append(reg)
    transforms.append(pca)
    

print('mean squared error: ' + str(sum(mse_data) / len(mse_data)))
print('r-squared: ' + str(sum(rsq_data) / len(rsq_data)))   
    
```
